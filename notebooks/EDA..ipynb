```python
## data preprocessing

df=df.drop(['sl_no','salary'],axis=1)
```

```python
df.head()
```

```python
## convert categorical columns into numerical

df['ssc_b'].unique()
```

```python
df['ssc_b']=df['ssc_b'].map({'Central':1,'Others':0})
```

```python
df.head()
```

```python
df['hsc_b'].unique()
```

```python
df['hsc_b']=df['hsc_b'].map({'Central':1,'Others':0})
```

```python
df.head()
```

```python
df['hsc_s'].unique()
```

```python
df['hsc_s']=df['hsc_s'].map({'Science':2,'Commerce':1, 'Arts':0})
```

```python
df.head()
```

```python
df['degree_t'].unique()
```

```python
df['degree_t']=df['degree_t'].map({'Sci&Tech':2,'Comm&Mgmt':1, 'Others':0})
```

```python
df.head()
```

```python
df['specialisation'].unique()
```

```python
df['specialisation']=df['specialisation'].map({'Mkt&HR':0,'Mkt&Fin':1})
```

```python
df.head()
```

```python
df['workex'].unique()
```

```python
df['workex']=df['workex'].map({'No':0,'Yes':1})
```

```python
df.head()
```

```python
df['status'].unique()
```

```python
df['status']=df['status'].map({ 'Not Placed':0,'Placed':1})
```

```python
df.head()
```

```python
## split data into dependent and independent variables

X=df.drop(['status'],axis=1)
```

```python
y=df['status']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=40)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
## Model Training

lr=LogisticRegression()
lr.fit(X_train,y_train)

knn=KNeighborsClassifier()
knn.fit(X_train,y_train)


dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)

rf=RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
## Prediction

y_pred_1=lr.predict(X_test)

y_pred_2=knn.predict(X_test)

y_pred_3=dt.predict(X_test)

y_pred_4=rf.predict(X_test)
```

```python
## Evaluting the model

from sklearn.metrics import accuracy_score

score_1=accuracy_score(y_test,y_pred_1)

score_2=accuracy_score(y_test,y_pred_2)
score_3=accuracy_score(y_test,y_pred_3)
score_4=accuracy_score(y_test,y_pred_4)
```

```python
final_score=pd.DataFrame({'Models':['LR','KNN','DT','RF'],
                         'Accuracy':[score_1*100,
                                     score_2*100,
                                      score_3*100,
                                      score_4*100]})
```
